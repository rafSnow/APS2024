```python
# Itera sobre as 10 postagens mais populares na subreddit "python"
for submission in reddit.subreddit('all').search('desmatamento', sort='relevance', time_filter='all'):
    print(submission.title)
    print(submission.display_name)
# - sort
    #'relevance': Ordena por relevância (o padrão).
    #'hot': Ordena por "hotness" (quente).
    #'new': Ordena por mais recentes.
    #'top': Ordena por mais votados.
    #'comments': Ordena por mais comentados.
    #'controversial': Ordena por mais controversos.
    
# - time_filter
    #'all': Todas as postagens.
    #'day': Postagens do último dia.
    #'hour': Postagens da última hora.
    #'month': Postagens do último mês.
    #'week': Postagens da última semana.
    #'year': Postagens do último ano.
```

```python
# Obtém as 10 postagens mais populares e as mais recentes
top_post = reddit.subreddit("python").top(limit=10)
new_post = reddit.subreddit("python").new(limit=10)
```

```python
# Itera sobre as postagens mais populares e imprime algumas informações
for post in top_post:
    print ("Titulo -", post.title)
    print("ID -", post.id)
    print("Autor -", post.author)
    print("URL -", post.url)
    print("Score -", post.score)
    print("Quantidades de Comentarios -", post.num_comments)
    print("Criado -", post.created_utc)
    print("\n")
```
